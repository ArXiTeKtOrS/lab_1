### Формула Эйлера (индивидуальное задание):

Формула Эйлера для комплексных чисел:

$$
e^{i\varphi} = \cos\varphi + i\sin\varphi
$$

Где:
- $e$ — основание натурального логарифма,
- $i$ — мнимая единица,
- $\varphi$ — вещественное число (угол в радианах).
