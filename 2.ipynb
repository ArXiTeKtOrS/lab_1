```python
import os

with open("file.txt", "w") as f:
    f.write("Первая строка\n")
    f.write("Вторая строка\n")
```

```python
with open ("file.txt", "r") as f:
    content = f.read()
    print("Содержимое файла:\n", content)

print("Файл существует:", os.path.exists("file.txt"))
```

```python
os.remove("file.txt")
print("Файл удалён")
```
